```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

control_rate = 0.10
variant_rate = 0.12

sims = 10000

control = np.random.binomial(1000, control_rate, sims)/1000
variant = np.random.binomial(1000, variant_rate, sims)/1000

diff = variant - control

print("Probability variant better:", np.mean(diff > 0))

plt.hist(diff, bins=40)
plt.axvline(0)
plt.title("Monte Carlo Simulation")
plt.show()
```
