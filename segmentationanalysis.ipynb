```python
import numpy as np
import pandas as pd

np.random.seed(42)

N = 10000

df = pd.DataFrame({
    "group": np.random.choice(["control","variant"], N),
    "device": np.random.choice(["mobile","desktop"], N)
})

df["converted"] = np.where(
    (df["group"]=="variant") & (df["device"]=="mobile"),
    np.random.binomial(1,0.14,N),
    np.random.binomial(1,0.10,N)
)

seg = df.groupby(["device","group"])["converted"].mean().unstack()
seg["lift"] = (seg["variant"] - seg["control"]) / seg["control"]

seg
```
